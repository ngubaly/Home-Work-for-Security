```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks, filtfilt, butter, peak_prominences
from scipy.spatial.transform import Rotation
from scipy.integrate import cumulative_trapezoid

column_names = ['Time stamp', 'ax', 'ay', 'az', 'gx','gy','gz','qw', 'qx', 'qy','qz']
data_new = pd.read_csv('/content/z_upward_2.csv', header = None, names = column_names)
data_new.head()
# Full-scale resolution of Data:
# Accelerometer: 4g; Gyroscope: 2000 dps
# Clock period: 10ms (f_clk = 100 Hz) --> 1s interval will get 100 experiments
# metric_value = raw_value * FSR / 2^15 - 1
```

## Filter function

```python
fs = 200
cutoff = 20
order = 4

def lowpass_filter(signal, cutoff, fs, order):
  nyquist = 0.5 * fs
  normalize_cutoff = cutoff / nyquist
  b, a = butter(order, normalize_cutoff, btype='low', analog=False)
  y = filtfilt(b, a, signal)
  return y

# Distinguish two type of frequency:
# + Frequency of signal (f):   Cycles per Second
# + Frequency of Sampling process (fs): Samples per Second
# The derived quantity: fs/f = Samples per Cycle (Theoretical minium is at least 2) --> fs/f >= 2 --> fs>= f/2 (Nyquist frequency)
```

## Check the the lowpass filter

```python
# Filter the acceleration
data_new['ax_filtered'] = lowpass_filter(data_new['ax'], cutoff, fs, order)
```

# **Estimate the velocity**: This work is implemented by using the inegral to proces

## This work will estimate the angle of IMU by using the acceleration data and by gyro data. Then using the complementary filter to get the accuracy angle

```python
# Constants
g = 9.81
fs = 200
dt = 1/fs # 5ms
# Create the data array and timestamp array
acc_data_filtered = data_new[['ax_filtered', 'ay_filtered', 'az_filtered']].values
# gyro_data_filtered = data_new[['gx_filtered', 'gy_filtered', 'gz_filtered']].values
gyro_data_filtered = data_new[['gx', 'gy', 'gz']].values
quaternion_data = data_new[['qw_filtered', 'qx_filtered', 'qy_filtered', 'qz_filtered']].values
time_stamp = np.arange(len(data_new))
```

### Estimate the angle from the acceleration data using trigonometry: we can only estimate the roll and pitch angle (this pair called as tilt)

#### Estimate angle by the acceleration data
The navigation equation in the acceleration form:
$
\begin{equation}
  \mathbf{a}^g = \mathbf{R}\cdot\mathbf{a}^s + \mathbf{g}^g
\end{equation}
$

With the global frame has the direction of each axis is:
*  x: don't care
*  y: don't care
*  z: upward
so the coordinate of the gravity vector in the global frame is: $ \mathbf{g}^g = [0, 0, -9.81]^T $

In the initial state, we can assume that the sensor is stationary or nearly so, i.e. the acceleration in the global frame is equal zero, $\mathbf{a}^g =0$ . Then the accelerometer measures the specific force is the reaction to gravity, thus in the body frame:
$\mathbf{a}^s_0 = \mathbf{R}^T_0 \cdot [0,0,9.81]^T$ or equivalent to
$\mathbf{R}_0\cdot\mathbf{a}^s_0 = [0,0,9.81]^T$

Normalize two vectors in two sides we have

$\mathbf{R}_0\cdot\mathbf{a}^s_{0,nor} = [0,0,1]^T$

Thus row of $\mathbf{R}$ can construct aligns with the gravity direction, with yaw arbitrarily set. Practical approach for construc $\mathbf{R}_0$:


*   Set the third column of $\mathbf{R}_0$ as $ \mathbf{u} =\mathbf{a}^s_{0,nor}$
*   Choose the first column as $\mathbf{v} = normalize(cross[1,0,0], u))$
* Set the second column as $\mathbf{w} = cross(u,v)$

Next, we will update the orientation the rotation matrix by using the gyro data:

The propagation equation of Direction cosine matrix:
$\mathbf{R}_k = \mathbf{R}_{k-1}\cdot \mathbf{R}_{\Delta}$,

where as the differential matrix $ \mathbf{R}_{\Delta}$ is the icremental rotation over $dt = 5ms $. For small $dt$, it can approximated as the skew-matrix
$\mathbf{R}_{\Delta} \approx
I + skew(\mathbf{\omega}_k\cdot dt)$.
In which, the skew matrix has the elements are the gyro data on each axis:
$\mathbf{skew}(\boldsymbol{\omega}) = \begin{bmatrix} 0 & -\omega_z & \omega_y \\ \omega_z & 0 & -\omega_x \\ -\omega_y & \omega_x & 0 \end{bmatrix}$
For better accuracy, use the exact rotation (using Taylor series)

$\mathbf{R}_{\Delta} \approx
\exp{(skew(\mathbf{\omega}_k\cdot dt))}$


```python
# Accel data processing
## Normalize some first initial acceleration datas
N_init = 200 # number of sample in the initial time
acc_data_initial = np.mean(acc_data_filtered[: N_init], axis=0)
acc_data_initial_normalize = acc_data_initial / np.linalg.norm(acc_data_initial)
## Initial the rotation matrix
u = acc_data_initial_normalize
v = np.cross([1,0,0], u) if np.abs(np.dot([1,0,0], u)) < 0.99 else np.cross([0,1,0], u)   # check if u parallel with the basis vector
v = v / np.linalg.norm(v)
w = np.cross(u, v)
R_0 = np.array([v,w,u]).T
R_list = [R_0]   # array for update rotation matrix
# Propagation the DCM
for k in range(1, len(time_stamp)):
  omega = gyro_data_filtered[k]
  skew = np.array([[0, -omega[1], omega[2]], [omega[1], 0, -omega[0]], [-omega[2], omega[0], 0]])
  R = R_list[-1] @ (np.eye(3) + skew * dt)
  R_list.append(R)
accel_data_global = np.array([R @ a for R, a in zip(R_list, acc_data_filtered)])
a_linear_dcm = accel_data_global - [0, 0, -g]
v_dcm = cumulative_trapezoid(a_linear_dcm, time_stamp, axis=0, initial=0)
v_dcm[:, 2]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(time_stamp, v_dcm[:, 0], label='vx', color = 'red')
plt.plot(time_stamp, v_dcm[:, 1], label='vy', color = 'blue')
plt.plot(time_stamp, v_dcm[:, 2], label='vz', color = 'green')
plt.title('Velocity vector')
plt.xlabel('Index')
plt.ylabel('Velocity')
# plt.ylim([-10,10])
plt.legend()
plt.grid(True)
plt.show()
```

### Estimate the velocity by using the quaternion

#### The quaternion $\mathbf{q}_k = [w, x, y, z]$ at each timestamp represents the rotation from body to global frame. Convert it to a rotation matrix:

$\mathbf{R}_k = \begin{bmatrix} 1 - 2(y^2 + z^2) & 2(xy - wz) & 2(xz + wy) \\ 2(xy + wz) & 1 - 2(x^2 + z^2) & 2(yz - wx) \\ 2(xz - wy) & 2(yz + wx) & 1 - 2(x^2 + y^2) \end{bmatrix} \
$

Ensure $\mathbf{q}_k$ is normalized $(w^2 + x^2 + y^2 + z^2 = 1)$; normalize if necessary.

```python
R_list_quater = [Rotation.from_quat(q).as_matrix() for q in quaternion_data ]
a_global_quater = np.array([R @ a for R, a in zip(R_list_quater, acc_data_filtered)])
a_linear_quater = a_global_quater - [0, 0, -g]
v_quater = cumulative_trapezoid(a_linear_quater, time_stamp, axis=0, initial=0)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(time_stamp, v_quater[:, 0], label='vx_quat', color = 'red')
plt.plot(time_stamp, v_dcm[:, 0], label='vx_dcm', color = 'blue')
plt.title('Velocity vector')
plt.xlabel('Index')
plt.ylabel('Velocity')
# plt.ylim([-10,10])
plt.legend()
plt.grid(True)
plt.show()
```

## Using the Complementary for DCM

```python
alpha = 0.99
def get_angle_from_accel(acc):
  roll = np.arctan2(acc[1], acc[2])
  pitch = np.arctan2(-acc[0], np.sqrt(acc[1]**2 + acc[2]**2))
  return roll, pitch

roll, pitch = get_angle_from_accel(acc_data_filtered[0])
yaw = 0;
angles = [(roll, pitch, yaw)]

# Apply the complementary filter
for k in range(1, len(time_stamp)):
  roll_gyro = angles[-1][0] + gyro_data_filtered[k][0] * dt
  pitch_gyro = angles[-1][0] + gyro_data_filtered[k][1] * dt
  yaw_gyro = angles[-1][2] + gyro_data_filtered[k][2] * dt

  # angle from accel
  roll_accel, pitch_accel = get_angle_from_accel(acc_data_filtered[k])
  # complementary filter
  roll = alpha * roll_gyro + (1 - alpha) * roll_accel
  pitch = alpha * pitch_gyro + (1 - alpha) * pitch_accel
  yaw = yaw_gyro
  angles.append((roll, pitch, yaw))

def angles_to_dcm (roll, pitch, yaw):
  cr, sr = np.cos(roll), np.sin(roll)
  cp, sp = np.cos(pitch), np.sin(pitch)
  cy, sy = np.cos(yaw), np.sin(yaw)
  return np.array([[cy*cp, cy*sp*sr - sy*cr, cy*sp*cr + sy*sr],
                   [sy*cp, sy*sp*sr + cy*cr, sy*sp*cr - cy*sr],
                   [-sp, cp*sr, cp*cr]])
R_list_comp = [angles_to_dcm(roll, pitch, yaw) for roll, pitch, yaw in angles]
a_global_comp = np.array([R @ a for R, a in zip(R_list_comp, acc_data_filtered)])
a_linear_comp = a_global_comp - [0, 0, -g]
v_comp = cumulative_trapezoid(a_linear_comp, time_stamp, axis=0, initial=0)

```

```python
plt.figure(figsize=(10, 6))
plt.plot(time_stamp, v_quater[:, 2], label='vx_quat', color = 'red')
plt.plot(time_stamp, v_dcm[:, 2], label='vx_dcm', color = 'blue')
plt.plot(time_stamp, v_comp[:, 2], label='vx_comp', color = 'green')
plt.title('Velocity vector')
plt.xlabel('Index')
plt.ylabel('Velocity')
# plt.ylim([-10,10])
plt.legend()
plt.grid(True)
plt.show()
```

## Using the propagation of quaternion


```python
def get_initial_quat(acc):
  pitch = np.arctan2(-acc[0], np.sqrt(acc[1]**2 + acc[2]**2))
  roll = np.arctan2(acc[1], acc[2])
  yaw = 0
  return Rotation.from_euler('xyz', [roll, pitch, yaw]).as_quat()

q = get_initial_quat(acc_data_filtered[0])
quat_list = [q]
# Cp nht quaternion, m bo du ng vi sensor-frame
for k in range(1, len(time_stamp)):
    omega = gyro_data_filtered[k]  # [wx, wy, wz], m bo du ng
    theta = np.linalg.norm(omega) * dt
    if theta > 0:
        axis = omega / (theta / dt)
        dq = np.array([np.cos(theta / 2), *(axis * np.sin(theta / 2))])
    else:
        dq = np.array([1, 0, 0, 0])
    q = (Rotation.from_quat(q) * Rotation.from_quat(dq)).as_quat()
    q /= np.linalg.norm(q)  # Chun ha
    quat_list.append(q)

# Chuyn sang ma trn xoay, tnh vn tc
R_list = [Rotation.from_quat(q).as_matrix() for q in quat_list]
a_global = np.array([R @ a for R, a in zip(R_list, acc_data_filtered)])
a_linear = a_global - [0, 0, -9.81]  # Tr trng lc, m bo global Z ln
v_quat_propa = cumulative_trapezoid(a_linear, time_stamp, axis=0, initial=0)

plt.figure(figsize=(10, 6))
plt.plot(time_stamp, v_quater[:, 0], label='vx_quat', color = 'red')
plt.plot(time_stamp, v_quat_propa[:, 0], label='vx_quat_propa', color = 'blue')
plt.title('Velocity vector')
plt.xlabel('Index')
plt.ylabel('Velocity')
# plt.ylim([-10,10])
plt.legend()
plt.grid(True)
plt.show()
```
